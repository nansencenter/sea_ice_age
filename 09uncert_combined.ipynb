```python
import os.path

from tqdm import tqdm

from lmsiage.uncertainty import ComputeAgeUncertainty
from lmsiage.zarr_index_tools import cleanup_missing_files, files_missing_array, update_index_for_dir
cleanup_missing_files()
```

```python
sia_dir = 'NERSC_arctic25km_sea_ice_age_v2p1/zarr'
mesh_dir = f'{sia_dir}/mesh'
age_dir = f'{sia_dir}/age'
unc_dir = f'{sia_dir}/uncert'

update_index_for_dir(unc_dir)
```

```python
unproc_unc_files = files_missing_array('unc_age')
unproc_unc_files = [f for f in unproc_unc_files if os.path.basename(f).split('_')[1].split('.')[0] > '19910914']
unproc_unc_files[0], unproc_unc_files[-1], len(unproc_unc_files)
```

```python
cau = ComputeAgeUncertainty(mesh_dir, age_dir, unc_dir)
for unc_file in tqdm(unproc_unc_files):
    status = cau.proc_unc_file(unc_file)
```
