```python
import glob
from tqdm.notebook import tqdm
from lmsiage.grid_age import ExportNetcdf
```

```python
lag_dir = 'NERSC_arctic25km_sea_ice_age_v2p1/zarr'
grid_dir = f'{lag_dir}/grid'
dst_root_dir = f'{lag_dir}/cmems'
osisaf_sic_dir = '/Volumes/sim/data/OSISAF_ice_conc_CDR_v3p0'

age_grd_files = sorted(glob.glob(f'{grid_dir}/*/grid_????????.zip'))
export_netcdf = ExportNetcdf(grid_dir, dst_root_dir, osisaf_sic_dir, 
                             file_prefix='nersc_siage_arctic65_', 
                             sia_name='siage',
                             dataset_doi=None,
                             step=3, 
                             export_conc=False,
                             export_lonlat=True,
                             monthly=False, 
                             force=True)
export_netcdf.set_attributes()



for age_grd_file in tqdm(age_grd_files[:5]):
    export_netcdf(age_grd_file)

for age_grd_file in tqdm(age_grd_files[-5:]):
    export_netcdf(age_grd_file)
```
